```python
def PAO(x): return 1 + sqrt(x^4 - x^2 + 1) + sqrt(x^4 + x^2)
```

```python
def PBO(x): return 1 + sqrt(x^4 + x^2 - 2*x + 1) + sqrt(x^4 + x^2)
```

```python
def r(x): return PAO(x)/PBO(x)
```

```python
plot(PAO(x),(x,0,5),legend_label='PAO(x)')
```

```python
plot(PBO(x),(x,0,5),legend_label='PBO(x)')
```

```python
plot(r(x),(x,0,5),legend_label='r(x)')
```
